```python
import asyncio
import json
import sys
from pathlib import Path

from dotenv import load_dotenv

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "src").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

load_dotenv(PROJECT_ROOT / ".env")

from src.core.providers.lmstudio_provider import LMStudioProvider
from src.planning.planner import Planner
from src.planning.scene_detector import SceneDetector
from src.planning.gap_analyzer import GapAnalyzer

LM_STUDIO_URL = "http://localhost:1234/v1"

# English text models.list() English text ID,English text
import openai

client = openai.OpenAI(base_url=LM_STUDIO_URL, api_key="lm-studio")
try:
    available_models = [m.id for m in client.models.list().data]
except Exception as e:
    available_models = []
    print(f"WARNING English text LM Studio models.list(): {type(e).__name__}: {e}")

print("English text LM Studio English text:", available_models)

# English text(English text)
MODEL_ID = available_models[0] if available_models else "gemma-4-e4b-it"
print(f"English text: {MODEL_ID}")

provider = LMStudioProvider(model=MODEL_ID, base_url=LM_STUDIO_URL)
planner = Planner(model_provider=provider)

planning_results = {}
planning_summary_rows = []
```

## Cell 3 - English text Planning(TC-01 ~ TC-04)

English text Planning English text,English text, English text, English text, English text expected critical English text.

```python
planning_results = {}
planning_summary_rows = []

sd = SceneDetector()
ga = GapAnalyzer(sd)

for tc in TEST_CASES[:4]:
    print(f"\n{'=' * 60}")
    print(f"[{tc['id']}] {tc['label']}")
    print(f"English text Prompt: {tc['prompt']}")

    try:
        result = await planner.plan(tc["prompt"])
        planning_results[tc["id"]] = result

        print(f"English text: {result.scene}")
        print(f"English text: {result.detectedIntent}")
        print(f"English text: {result.refinedRequirements}")
        print(f"\nEnglish text ({len(result.missingFields)} English text):")

        for mf in result.missingFields:
            marker = "Required" if mf.importance == "critical" else ("Recommended" if mf.importance == "recommended" else "Optional")
            default_str = f"(English text: {mf.defaultValue})" if mf.defaultValue else ""
            print(f"  {marker} [{mf.field}] {mf.question} {default_str}")

        questions = ga.generate_questions(result.missingFields)
        print(f"\nEnglish text:\n{questions}")

        actual_criticals = [mf.field for mf in result.missingFields if mf.importance == "critical"]
        for expected in tc.get("expected_criticals", []):
            status = "PASS" if expected in actual_criticals else "FAIL"
            print(f"  {status} English text critical English text '{expected}': {'English text' if expected in actual_criticals else 'English text'}")

        planning_summary_rows.append(
            {
                "English text ID": tc["id"],
                "English text": tc["label"],
                "English text": result.scene,
                "English text(English text 40 English text)": result.detectedIntent[:40],
                "critical English text": sum(1 for mf in result.missingFields if mf.importance == "critical"),
                "recommended English text": sum(1 for mf in result.missingFields if mf.importance == "recommended"),
                "English text": "PASS English text",
            }
        )

    except Exception as e:
        print(f"FAIL [{tc['id']}] English text: {type(e).__name__}: {e}")
        print("  -> IntentAnalyzer English text,English text LM Studio English text")
        planning_summary_rows.append(
            {
                "English text ID": tc["id"],
                "English text": tc["label"],
                "English text": "N/A",
                "English text(English text 40 English text)": "N/A",
                "critical English text": None,
                "recommended English text": None,
                "English text": f"FAIL English text: {type(e).__name__}",
            }
        )

print(f"\n{'=' * 60}")
```

## Cell 4 - English text(TC-05)

English text `plan_with_supplement`,English text, English text.

```python
tc = TEST_CASES[4]  # TC-05

try:
    print(f"[{tc['id']}] {tc['label']}")
    print(f"English text Prompt: {tc['prompt']}\n")

    # Step 1: English text plan(English text)
    result_before = await planner.plan(tc["prompt"])
    criticals_before = [mf.field for mf in result_before.missingFields if mf.importance == "critical"]
    print(f"English text - English text {len(result_before.missingFields)} English text,critical: {criticals_before}")

    # Step 2: English text plan
    result_after = await planner.plan_with_supplement(tc["prompt"], tc["supplements"])
    criticals_after = [mf.field for mf in result_after.missingFields if mf.importance == "critical"]
    print(f"English text - English text {len(result_after.missingFields)} English text,critical: {criticals_after}")
    print(f"English text: {result_after.refinedRequirements}")

    # English text:English text critical English text
    improvement = len(result_before.missingFields) - len(result_after.missingFields)
    status = "PASS" if improvement >= 0 else "WARNING"
    print(f"\n{status} English text {len(result_before.missingFields)} -> {len(result_after.missingFields)}(English text {improvement} English text)")

except Exception as e:
    print(f"FAIL TC-05 English text: {type(e).__name__}: {e}")
```

```python
import pandas as pd
from IPython.display import Markdown, display

summary_df = pd.DataFrame(
    planning_summary_rows,
    columns=["English text ID", "English text", "English text", "English text(English text 40 English text)", "critical English text", "recommended English text", "English text"],
)

display(summary_df)

success_count = int(summary_df["English text"].astype(str).str.startswith("PASS").sum()) if not summary_df.empty else 0
total_count = len(summary_df)
critical_values = pd.to_numeric(summary_df["critical English text"], errors="coerce").dropna()
avg_critical = critical_values.mean() if not critical_values.empty else None

conclusion_lines = [
    "### Planning Engine English text",
    f"- TC-01 ~ TC-04 English text {total_count} English text,English text {success_count} English text.",
    "- English text, English text, React/TypeScript English text FastAPI English text.",
    "- English text, English text vibe coding prompt English text.",
]
if avg_critical is not None:
    conclusion_lines.append(f"- English text critical English text {avg_critical:.2f},English text prompt English text.")
else:
    conclusion_lines.append("- English text critical English text,English text LM Studio English text.")

display(Markdown("\n".join(conclusion_lines)))
```
